```python
from lib.cartpoles import Cart, Pole, DCMotor
from lib.colors import Colors
import numpy as np
from matplotlib import pyplot as plt

dt = 0.01
g = 9.81

cart = Cart(0.5, 0.05, 0, -0.8, 0.8, Colors.red,
    DCMotor(-12, 12, 0.05, 0.5, 0.05, 0.01, 0.05, Colors.black),
    Pole(0.2, np.radians(10), 0.2, 0.005, Colors.green, 
        Pole(0.2, np.radians(5), 0.15, 0.005, Colors.blue, 
            Pole(0.2, np.radians(15), 0.10, 0.005, Colors.purple, None
            )
        )
    )
)

for i in range(100000):
    cart.update(dt, 0, g)

x = cart.state["x"]
d_x = cart.state["d_x"]
dd_x = cart.state["dd_x"]

ax = plt.figure().add_subplot(projection="3d")
ax.plot(x, d_x, dd_x, "b")
```
