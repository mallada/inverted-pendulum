```python
from eom import calculate_equations_of_motions, save_equations_of_motions
```

```python
def calculate_display_save(n_poles: int = 1):
    assert n_poles > 0
    sols, pure_sols = calculate_equations_of_motions(n_poles)
    print("Solutions:")
    for i, sol in enumerate(sols):
        name = "tau" if i == 0 else f"dd_theta_{i}"
        print(f"{name}:")
        display(sol)
    save_equations_of_motions(n_poles, pure_sols)
```

```python
calculate_display_save(1)
```

```python
calculate_display_save(2)
```
